```python
import math
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python

def probability(x_new, x_prev, y, t):
    p = 1.
    y_new = y(x_new)
    y_prev = y(x_prev)
    if y_new - y_prev > 0:
        p = math.e**(-(y_new - y_prev)/t)
    return p

def draw(x_min, y, t):
    fig, ax = plt.subplots()
    x = np.linspace(-15, 10, 100)
    yf = y(x)
    ax.plot(x, yf, linewidth=2.0)
    ax.axvline(x=x_min, color = 'red')
    plt.text (-15, 300, 't = ' + str(round(t,2)))
    plt.show()

x_min = 0.
y = lambda x: 100*(0.01*(x+7)**2 + np.sin(x))
t = 100.
draw(x_min, y, t)
x_list = []
while t > 0:
    x = x_min + random.uniform(-10, 10)
    x_min = random.choices([x_min, x], weights=[1 - probability(x, x_min, y, t), probability(x, x_min, y, t)])[0]
    x_list.append(x_min)
    t += -0.2 
    draw(x_min, y, t)

x_list.sort()
print(x_min)
print(x_list[0])

```
